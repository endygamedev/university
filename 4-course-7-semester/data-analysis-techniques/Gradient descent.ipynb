## Метод градиентного спуска

### Импорт модулей

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

### Данные

```python
l = 1000
x =  np.arange(l)/l
y = 10*x + np.random.standard_normal(l)
```

```python
plt.scatter(x,y,s=2)
plt.show()
```

### Градиент

```python
def grad(theta: np.array, x: np.array, y: np.array) -> np.array:
    return np.array((
                    -1/l*np.sum([y[i]-theta[0]-theta[1]*x[i] for i in range(len(x))]),
                    -1/l*np.sum(([(y[i]-theta[0]-theta[1]*x[i])*x[i] for i in range(len(x))]))))
```

### Метод

```python
def gradient_descent(x: np.array, y: np.array, *,
                     theta_0=np.array((0,0)),
                     nu=0.1,
                     n=1000,
                     eps=10**-5) -> np.array:
    i = 0
    prev = theta_0 + 1000
    theta = theta_0
    while i < n or np.linalg.norm(theta - prev) <= eps:
        prev = theta
        theta = theta - nu * grad(theta, x, y)
        i += 1
    return theta
```

```python
theta = gradient_descent(x, y)
```

```python
plt.plot(x, theta[0]+theta[1]*x)
plt.scatter(x,y,s=2)
plt.show()
```
